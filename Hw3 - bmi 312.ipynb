```python
clinical_short = clinical[['patient', 'year_of_birth', 'gender', 'race', 'definition', 'tumor_stage', 'age_at_diagnosis', 'vital_status', 'days_to_death', 'days_to_last_follow_up']]
```

```python
clinical_short
```

```python
conversion = clinical_short['age_at_diagnosis']/365 # Convert “ age_at_diagnosis ” to years, round to the nearest tenths and overwrite the original values.
conversion2 = clinical_short['days_to_death']/12 # Convert “ days_to_death ” to months, round to the nearest tenths and overwrite the original values.
conversion3 = clinical_short['days_to_last_follow_up']/12 # Convert “ days_to_last_follow_up ” to months ” to months, round to the nearest tenths and overwrite the original values.
conversion
```

```python
conversion2
```

```python
conversion3
```

```python
clinical_short['age_at_diagnosis'] = conversion
clinical_short['days_to_death'] = conversion2
clinical_short['days_to_last_follow_up'] = conversion3
clinical_short
```

```python
clinical_short = clinical_short.round({"age_at_diagnosis":1, "days_to_death":1, "days_to_last_follow_up":1})
clinical_short
```

```python
clinical_short = clinical_short.rename(columns={"days_to_death" : "months_to_death", "days_to_last_follow_up" : "months_to_last_follow_up"})
clinical_short
```

```python
clinical_short.drop(clinical_short.index[clinical_short['definition'] == 'Normal'], inplace=True) 
clinical_short # Remove patients with non cancer diagnosis from “ clinical_short"
```

```python
# Save “ clinical_short ” in a tab delimited text file.
clinical_short.to_csv('/Users/hannahkaufman/Downloads/Data/clinical_short.txt', sep='\t', index=False) 
```
